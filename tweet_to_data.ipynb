```python
import tweepy as tw
from tqdm import tqdm
import pandas as pd
import os
import datetime
import re
```

```python
api_key = ''
api_secret_key = ''
bearer_token = ''
access_token = ''
access_token_secret = ''
```

```python
auth = tw.OAuthHandler(api_key, api_secret_key)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
# Define the search term and the date_since date as variables
search_words = "woolworths australia"

# Collect tweets
tweets = tw.Cursor(
              api.search,
              tweet_mode='extended',
              retweeted_status='full_text',
              q=search_words,
              lang="en",
                    ).items(10000)

# tweets = tw.Cursor(
#               api.search,
#               tweet_mode='extended',
#               retweeted_status='full_text',
#               q=search_words,
#               lang="en",
#               since=date_since
#                     ).items(NUM_TWEETS)
```

```python
for tweet in tqdm(tweets):
    t = tweet
    d = {}
    d['author_follower_count'] = [tweet.author.followers_count]
    d['author_friends_count'] = [tweet.author.friends_count]
    d['author_listed_count'] = [tweet.author.listed_count]
    d['author_favourites_count'] = [tweet.author.favourites_count]
    d['user'] = [tweet.author.name]
    d['author_screen_name'] = [tweet.author.screen_name]
    d['author_location'] = [tweet.author.location]
    d['author_created_at'] = [tweet.author.created_at]
    
    try:
        d['full_text'] = [re.sub('&amp;','and', re.sub('https:\/\/t\.co\/.{10}','',tweet.retweeted_status.full_text))]
    except AttributeError:  # Not a Retweet
        d['full_text'] = [re.sub('&amp;','and', re.sub('https:\/\/t\.co\/.{10}','', tweet.full_text))]
    d['length'] = [len(d['full_text'])]
    d['created_at'] = [tweet.created_at.strftime('%Y-%m-%d %H:%M')]
    d['hashtags'] = [','.join([i['text'] for i in t.entities['hashtags']])]
    d['user_mentions'] = [','.join([i['name'] for i in tweet.entities['user_mentions']])]
    d['num_user_mentions'] = [len(tweet.entities['user_mentions'])]
    d['source'] = [tweet.source]
    d['retweet_count'] = [tweet.retweet_count]
    d['favorite_count'] = [tweet.favorite_count]
    d['geo'] = [tweet.geo]
    d['coordinates'] = [tweet.coordinates]
    d['place'] = [tweet.place]
    d['contributors'] = [tweet.contributors]
    d['search_term'] = ['costco australia']

    try:
        df = pd.concat([df, pd.DataFrame.from_dict(d)], ignore_index=True)
    except:
        df = pd.DataFrame.from_dict(d)
```
